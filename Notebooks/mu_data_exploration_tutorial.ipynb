```python
crispr_mu = crispr_mu[crispr_mu['gene'].obs.query('total_gene_umis  > 0').index.values.tolist(), :]
```

```python
 
filter_guides = 'NGFRAP1|GAGGACCGAGAAGAGTGACA|TGTCACTCTTCTCGGTCCTCA'
filter = (crispr_mu['guide'][: , filter_guides].to_df() > 0)[filter_guides].tolist()
```

```python
no_guide_presence_filter = (crispr_mu['guide'][: , filter_guides].to_df() == 0)[filter_guides].tolist()
```

```python
#Selecting just cells with a given guide 
crispr_mu['guide'][filter, : ]
```

```python
#USING TO CAPTURE TRANSCRIPTS
crispr_mu['gene'][filter, : ].to_df()
```

```python
crispr_mu['gene'].var.loc['ENSG00000166681']
```

```python
#using the gene name to capture the transcripts

Gene_name = 'BEX3'
ensembl = crispr_mu['gene'].var.query(f'symbol ==  "{Gene_name}" ') ['symbol'].index[0]
ensembl
```

```python
df_w_guide = crispr_mu['gene'][filter,  ensembl].to_df()

df_w_guide['has_guide'] = 'YES'
df_w_guide
```

```python
df_set_no_guide = crispr_mu['gene'][ no_guide_presence_filter,  ensembl].to_df().head(300)
df_set_no_guide['has_guide'] = 'NO'

df_set_no_guide
```
